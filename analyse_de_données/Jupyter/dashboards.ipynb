# Analyse des donn√©es

```python
import pandas as pd
import numpy as np
from scipy.stats import norm
import statistics
import matplotlib.pyplot as plt

data_df = pd.read_csv("final_data.csv")
dict = {'Unnamed: 0': 'number'}
data_df.rename(columns = dict, inplace = True)

plt.rcParams["figure.figsize"] = [8, 5.3]
plt.rcParams["figure.autolayout"] = True


def plot_training(column_name, column, start=False, end=False, step=False):

    """This function is used to visualize the data."""
    plt.hist(column, rwidth=0.8)
    if type(start) in (int, float):
        plt.xticks(np.arange(start, end, 1000000))
    plt.xlabel('Valeurs')
    #plt.xticks(np.arange(0, 1, 0.1))
    plt.ylabel('Quantit√©')
    plt.title(column_name)
    plt.legend(column_name)
    plt.savefig(column_name +'_figure.png')
    plt.show()

    
number = data_df["number"]
valence = data_df["valence"]
acousticness = data_df["acousticness"]
danceability = data_df["danceability"]
duration_ms = data_df["duration_ms"]
energy = data_df["energy"]
explicit = data_df["explicit"]
instrumentalness = data_df["instrumentalness"]
key = data_df["key"]
liveness = data_df["liveness"]
loudness = data_df["loudness"]
mode = data_df["mode"]
popularity = data_df["popularity"]
release_date = data_df["release_date"]
speechiness = data_df["speechiness"]
tempo = data_df["tempo"]
artists_pop = data_df["artists' popularity"]
genres_pop = data_df["genres' popularity"]
year_pop = data_df["year_popularity"]
top50 = data_df["top50"]

#data_df.groupby("release_date")["duration_ms"].sum().plot.bar(rot=45, legend= True)

plot_training("Valence", valence)
```

```python
plot_training("Acousticness", acousticness)
```

```python
plot_training("Danceability", danceability)
```

```python
#plot_training("Release date", release_date)
#plot_training("years' popularity", year_pop)
plt.figure(figsize=[8, 5.3])
plt.subplot(2,1,1 )
#plt.plot(release_date,number, label="Release date", c='red')
data_df.groupby("release_date")["number"].sum().plot()
plt.title("Release date")
plt.legend("Release date")
plt.savefig('release_date_figure.png')
plt.plot()
plt.show()

plt.subplot(2,1,2)
plt.title("years' popularity")
plt.legend("years' popularity")
data_df.groupby("release_date")["year_popularity"].sum().plot()
plt.savefig('year_popularity_figure.png')
plt.show()
```

```python


plt.figure(figsize=[8, 5.3])
plt.subplot(1,2,1)
plt.title("mode")
plt.legend("mode")
data_df.groupby("mode")["number"].sum().plot.pie(legend=True)
plt.savefig('mode_figure.png')
plt.show()

plt.subplot(1,2,2)
plt.title("explicit")
plt.legend("explicit")
data_df.groupby("explicit")["number"].sum().plot.pie(legend=True)
plt.savefig('explicit_figure.png')
plt.show()

```
