```python

import pandas as pd
df_clean_data = pd.read_csv("clean_data.csv")
#Ce code nettoie une table de données en supprimant les caractères superflus (e.g. crochets autour des listes, apostrophes...)
df_clean_data.artists = df_clean_data.artists.apply( lambda x: x.replace( "[", ""))
df_clean_data.artists = df_clean_data.artists.apply( lambda x: x.replace( "'", ""))
df_clean_data.artists = df_clean_data.artists.apply( lambda x: x.replace( "]", ""))
df_clean_data["artists genres"] = df_clean_data["artists genres"].apply( lambda x: x.replace( "[", ""))
df_clean_data["artists genres"] = df_clean_data["artists genres"].apply( lambda x: x.replace( "'", ""))
df_clean_data["artists genres"] = df_clean_data["artists genres"].apply( lambda x: x.replace( '"' , "" ))
df_clean_data["artists genres"] = df_clean_data["artists genres"].apply( lambda x: x.replace( "]", ""))
#supprimer les colonnes non utilisées
df_clean_data = df_clean_data.drop(["id", "Unnamed: 0" , "artists' followers" ], axis=1 )

#df_clean_data[df_clean_data["year_popularity"].isnull()] 

#supprimer les lignes dupliquées
df_clean_data = df_clean_data.drop_duplicates(subset=["name", "artists", "release_date"])

df_clean_data   
```

```python
#calcul des medianes
m_artists = df_clean_data["artists' popularity"].median()
m_genres = df_clean_data["genres' popularity"].median()
m_year = df_clean_data["year_popularity"].median()
m_artists
```

```python
#remplacer les NaN par la médiane de la colonne
df_clean_data["artists' popularity"] = df_clean_data["artists' popularity"].fillna(37.19047619047619)
df_clean_data["artists' popularity"]
```

```python
#remplacer les NaN par la médianne de la colonne
df_clean_data["genres' popularity"] = df_clean_data["genres' popularity"].fillna(m_genres)
df_clean_data["genres' popularity"]
```

```python
#remplacer les NaN par la médianne de la colonne
df_clean_data["year_popularity"] = df_clean_data["year_popularity"].fillna(m_year)
df_clean_data["year_popularity"]
```

```python
#trier les données en fonction de leur popularité
#df_clean_data = df_clean_data.set_index("popularity")
df_clean_data = df_clean_data.sort_index(axis = 0, ascending = False)

#ajouter une colonne top50
df_clean_data = df_clean_data.assign(top50=0)
df_clean_data.head(50)
```
