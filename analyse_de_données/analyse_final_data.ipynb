```python
# Supprime les doublons dans la table

import pandas as pd
df_clean01 = pd.read_csv("clean_data01.csv")
df_clean01
df_clean02 = df_clean01.drop_duplicates(subset=["name", "artists", "release_date"])
df_clean02 = df_clean02.drop(["Unnamed: 0"], axis=1 )
df_clean02
```

```python
m_artists = df_clean02["artists' popularity"].median()
m_genres = df_clean02["genres' popularity"].median()
m_year = df_clean02["year_popularity"].median()
m_artists
```

```python
m_genres
```

```python
m_year
```

```python
df_clean02["artists' popularity"] = df_clean02["artists' popularity"].fillna(37.19047619047619)
df_clean02["artists' popularity"]
```

```python
df_clean02["genres' popularity"] = df_clean02["genres' popularity"].fillna(m_genres)
df_clean02["genres' popularity"]
```

```python
df_clean02["year_popularity"] = df_clean02["year_popularity"].fillna(m_year)
df_clean02["year_popularity"]
```
